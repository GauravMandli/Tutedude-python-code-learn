```python
import pandas as pd
path="D:/Tutedude-py/Tutedude-python-code-learn/data analysis/data3.xlsx"
df=pd.read_excel(path)
print(df)
```

```python
# numeric to categorical
import numpy as np
bins=np.linspace(min(df["length"]),max(df["length"]),4)
```

```python
group_name=["low","medium","high"]
```

```python
df["length-binned"]=pd.cut(df["length"],bins,labels = group_name,include_lowest = True)
```
