```python
#distribution plot
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(0)
x=np.random.randn(100)
ax_1=sns.distplot(x)
```

```python
ax_2 = sns.displot(x=x, kind="kde", rug=True)

```

```python
ax_3=sns.displot(x=x, kind="kde", rug=False)
```

```python
from scipy.stats import norm
# 1 Histogram
ax_4=sns.histplot(x, stat="density", kde=True, color="skyblue")

# 2 Normal distribution fit
mu, sigma = norm.fit(x)
xmin, xmax = min(x), max(x)
x_vals = np.linspace(xmin, xmax, 100)
y_vals = norm.pdf(x_vals, mu, sigma)

# 3 Plot normal curve
plt.plot(x_vals, y_vals, color="black", label="Normal Fit")

plt.legend()
plt.show()
```

```python
ax_5=sns.histplot(
    y=x,          #  y use karo (this is the key)
    kde=True,
    stat="density",
    color="skyblue"
)
```

```python
sns.set_color_codes()
ax_6=sns.histplot(x, kde=True, stat="density", color="r")
```

```python
sns.set(style="dark")
ax_7=sns.histplot(x, kde=True, stat="density", color="g")
```
