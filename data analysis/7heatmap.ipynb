```python
flights= sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year",  values="passengers")
heat_map_4=sns.heatmap(flights_pivot)
```

```python
heat_map_5=sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")
```

```python
heat_map_5=sns.heatmap(flights_pivot, annot=True, fmt=".1f", cmap="coolwarm")
```

```python
heat_map_8=sns.heatmap(flights_pivot, cbar=False, cmap="coolwarm")
```
