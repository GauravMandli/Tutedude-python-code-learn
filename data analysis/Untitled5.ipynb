```python
import pandas as pd
path="D:/Tutedude-py/Tutedude-python-code-learn/data analysis/data3.xlsx"
df=pd.read_excel(path)
print(df)
```

```python
#boxplot
import seaborn as sns
sns.boxplot(x="length",y="E",data = df)
```

```python
#scatterplot
import matplotlib.pyplot as plt
x=df["length"]
y=df["E"]

plt.scatter(x,y)
plt.title("Scatterplot along the x and y axis")

plt.xlabel("length")
plt.ylabel("width")
```

```python
import numpy as np 
grp=df.groupby("A")
grp.aggregate(np.sum)
```

```python
grp=df.groupby(["A","B"])
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A","B"], as_index=False)
grp.aggregate(np.sum)
```
